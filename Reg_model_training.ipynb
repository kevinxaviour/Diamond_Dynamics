```python
X=df[['log_carat','cut_encoded', 'clarity_encoded', 'color_encoded','depth_ws']]
y=df['price']
ss=StandardScaler()
X_scaled=ss.fit_transform(X)
X_train_scaled,X_test_scaled,y_train,y_test=train_test_split(X_scaled,y,test_size=0.2,random_state=42)
```

```python
lr=LinearRegression()
lr.fit(X_train_scaled,y_train)
```

```python
y_train_pred=lr.predict(X_train_scaled)
y_test_pred=lr.predict(X_test_scaled)
```

```python
# --- Training Metrics ---
mae_train = mean_absolute_error(y_train, y_train_pred)
mse_train = mean_squared_error(y_train, y_train_pred)
rmse_train = np.sqrt(mse_train)
r2_train = r2_score(y_train, y_train_pred)

# --- Testing Metrics ---
mae_test = mean_absolute_error(y_test, y_test_pred)
mse_test = mean_squared_error(y_test, y_test_pred)
rmse_test = np.sqrt(mse_test)
r2_test = r2_score(y_test, y_test_pred)

# --- Display ---
print("Training Set Metrics:")
print(f"MAE  : {mae_train:.4f}")
print(f"MSE  : {mse_train:.4f}")
print(f"RMSE : {rmse_train:.4f}")
print(f"R²   : {r2_train:.4f}\n")

print("Testing Set Metrics:")
print(f"MAE  : {mae_test:.4f}")
print(f"MSE  : {mse_test:.4f}")
print(f"RMSE : {rmse_test:.4f}")
print(f"R²   : {r2_test:.4f}")
```

```python
sy=StandardScaler()
X_scaled=ss.fit_transform(X)
ys = y.values.reshape(-1, 1)
y_scaled=sy.fit_transform(ys)
X_train_scaled,X_test_scaled,y_train_scaled,y_test_scaled=train_test_split(X_scaled,y_scaled,test_size=0.2,random_state=42)
```

```python
model = Sequential([
    Dense(64, input_dim = X_train_scaled.shape[1], activation = 'relu'),
    Dense(32, activation = 'relu'),
    Dense(16, activation = 'relu'),
    Dense(1, activation = 'linear')
])
```

```python
model.compile(optimizer='adam', loss='mean_squared_error')
```

```python
model.fit(X_train_scaled, y_train_scaled, epochs=100, batch_size=32, validation_data=(X_test_scaled, y_test_scaled))
```

```python
y_test_actual=sy.inverse_transform(y_test_scaled)
test_loss = model.evaluate(X_test_scaled,y_test_scaled )
print(f"Test Loss: {test_loss:.4f}")
```

```python
# Predict on train and test scaled inputs
y_train_pred_scaled = model.predict(X_train_scaled)
y_test_pred_scaled  = model.predict(X_test_scaled)

# Inverse transform all
y_train_pred = sy.inverse_transform(y_train_pred_scaled)
y_test_pred  = sy.inverse_transform(y_test_pred_scaled)

y_train_actual = sy.inverse_transform(y_train_scaled)
y_test_actual  = sy.inverse_transform(y_test_scaled)
```

```python
def evaluate_regression(y_true, y_pred, name="Data"):
    mae  = mean_absolute_error(y_true, y_pred)
    mse  = mean_squared_error(y_true, y_pred)
    rmse = np.sqrt(mse)
    r2   = r2_score(y_true, y_pred)
    
    print(f"\n{name} Evaluation")
    print(f"MAE  : {mae:.4f}")
    print(f"MSE  : {mse:.4f}")
    print(f"RMSE : {rmse:.4f}")
    print(f"R²   : {r2:.4f}")
```

```python
evaluate_regression(y_train_actual, y_train_pred, name="Training Set")
```

```python
evaluate_regression(y_test_actual, y_test_pred, name="Testing Set")
```
